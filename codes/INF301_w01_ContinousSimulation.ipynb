## System Simulation

Simulation can be understood as the study of the **behavior of real systems** through the construction of models, which have features representing the behavior of the real system.

A system can be interpreted as a collection of items among which some functional relationship can be found or defined.

The term “simulation” is quite generic, as there are various forms and mechanisms used to represent system behavior.

## Fields of Application

The term “simulation” also has a broad scope in terms of its application field.
* Simulation can be applied in industry, public organizations, software/hardware representation, etc.

The benefits of system simulation, combined with the speed offered by computational tools, have been used as an auxiliary tool in solving various problems. This is justified by the fact that, with the use of a computer, a large number of events can be executed in a short period of time.

## Use of Simulation
Simulation is especially recommended in two cases:

* First, when solving a problem through experimentation is too costly or even impossible.

* Second, when problems are too complex for analytical treatment.

By using simulation, especially when stochastic characteristics are involved, systems can have their behavior represented more accurately and realistically.

## Benefits of Simulation
Factors that make the use of simulation techniques desirable include:

* Time: with a computer, it is possible to conduct experiments that would take years if performed on the real system;

* Cost: although computer simulation requires human resources and equipment, its cost is usually much lower than that of experiments on the real system;

* Impossibility of direct experimentation: in some cases, direct experimentation on the real system is not possible due to safety, time, access, or because the system does not yet exist;

* Visualization: computers offer tools that facilitate visualization of simulation results (charts, tables, etc.), as well as the system's state during model execution;

* Repetition: once built, a simulation model can be executed multiple times at very low cost;

* Interference: a model is much more flexible than the real system when it comes to implementing changes. This is particularly desirable when studying systems to generate decision-support information.

## Simulation Models

Simulation models can be regarded as descriptions of real systems.

Running simulation models on computers has the potential to produce more accurate results without interfering with the real system.

When analyzed statistically, these results can provide valuable information for decision-making aimed at solving problems.

Another relevant point in simulation is the order of event occurrence.
An event (a state change) is any occurrence that interferes with the system.

Since simulation deals with system behavior—i.e., the sequence of events—it is essential to track a time variable.

The time at which each event occurs is studied to evaluate whether it should happen at that moment, earlier, or later.

## Applications of Simulation

Simulation can be used to:

* Estimate the distribution of random variables;

* Test statistical hypotheses;

* Compare scenarios representing different solutions to a studied problem;

* Evaluate the behavior of an analytical solution;

* Assess real-time decision-making processes.

## Definition of System

For the purposes of this course, a system can be defined as a structured set of items (components or subsystems) among which a functional relationship can be found or defined.

The value of a component's attributes at a given time defines the model's state.

## Model Construction

In general, models are used to solve problems that would be too expensive to address experimentally or too complex for analytical treatment.

A model can be built to address a specific problem within a system.

There may be several different models for the same system, each one tailored to a particular problem.

Once a problem within a system that can be modeled is identified, an organized structure must be used to describe it.

## Computational Model

A simulation language is used to provide this structure and allow translation of the model into a format suitable for computer processing.

The computer is then used to execute the model and produce outputs that can be analyzed, guiding decisions aimed at solving the problem.

With the aid of computers, a large number of events can be simulated in a short time.

## Model and Simulation

A model representing a system allows us to estimate its future behavior. Running a model is called simulation.

It's important to emphasize that experiments are not the same as simulations.
While experiments involve making changes to the real system, simulations are conducted on a system's representation.

Moreover, experiments can be extremely expensive, justifying the construction of simulation models in many cases.

## Analysis

Practical applications of simulation include studying environmental changes, complex transformations in certain settings, or training (e.g., flight simulators).

Model simulation results are analyzed to understand new or unfamiliar scenarios, inform new rules, or gain experience for improving the model itself.

## Model Complexity

Models must incorporate real elements in a simple way, balancing cost and benefit.

Avoid creating models that are overly complex and hard to understand or manipulate.

Also avoid overly simple models that fail to capture the system's essential properties.

## System Design

A system model is rarely perfect in terms of accuracy, and there are two basic modeling strategies: top-down and bottom-up.

* In a top-down approach, modeling begins with high-abstraction modules, which are progressively refined.

* In a bottom-up approach, modeling starts at a low level of abstraction, with increasingly abstract structures added until the whole system is defined.

The challenge is that low-level structures are not always precisely defined (e.g., atomic-level phenomena).

The power and widespread acceptance of system simulation lie in the fact that it works and leads to satisfactory results.

## System Properties
The system input, x(t), represents all external influences and can be expressed as a vector with m real-valued variables:

x(t) = [x₁(t), x₂(t), ..., xₘ(t)]

All system effects can be represented by an output vector z(t), with n real-valued variables:

z(t) = [z₁(t), z₂(t), ..., zₙ(t)]

If the system depends only on the input, it is said to be of zero order.

There is also a state vector y(t) = [y₁(t), y₂(t), ..., yₚ(t)].

If the system also depends on the output, it is considered dynamic.

## Example 2

Consider a system in a factory where boxes arrive at a rate of one every 10 seconds. Each box has one of the following weights (w): 5, 10, or 15 kg. However, there are twice as many 5 kg and 15 kg boxes as there are 10 kg boxes.
How can this system be modeled and simulated?

<center>

image.png

</center>

Statistical Model:

$$
\begin{array}{|c|c|}
  \hline
  w & Pr[W = w] \\ \hline
  5 & 0.4 \\
  10 & 0.2 \\
  15 & 0.4 \\ \hline
\end{array}
$$

Note that the sum of all probabilities is 1: $\sum Pr[W=w] = 1$.


````
import numpy as np
import matplotlib.pyplot as plt # for the graphic

n = 1000
w = np.zeros(3)
Pexp = np.zeros(3)
for k in range(n):
   r = 10*np.random.random_sample()
   if r < 4:
      w[0]+=1
   elif r >= 4 and r < 6:
      w[1]+=1
   else:
      w[2]+=1

weight = np.array([5, 10, 15])

for i in range(len(w)):
   Pexp[i] = w[i]/n

Pt = np.array([0.4, 0.2, 0.4])
barWidth = 0.25

# Determining the central position for the bars
x1 = np.arange(len(weight))
x2 = [x - barWidth/2 for x in x1]
x3 = [x + barWidth/2 for x in x1]

#  Figure
fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = (8,6)
plt.rcParams["font.size"] = "12"
r1 = ax.bar(x2, Pt, width = 0.25, color = "blue", label="Theory")
r2 = ax.bar(x3, Pexp, width = 0.25, color = "darkred", label="Simulation")
ax.set_ylabel("Probability",fontsize=14)
ax.set_xlabel("Weights",fontsize=14)
ax.set_xticks(x1)
ax.set_xticklabels(weight)
plt.legend(prop={"size":14},frameon=False)
plt.show()
````

## Using Simulations to Solve Problems

In a problem-solving context, a simulation is considered valid for the specific context in which it is applied.

For example, if a researcher is studying rocket trajectories via simulation, this means the rocket should be used under circumstances that match those expected by the simulation designer.

Simulations for problem-solving generally allow parameter selection to alter scenarios and ensure that different system behaviors can be studied.

## Types of Problems

Simulation has been applied in many areas, such as:

* analysis of modern computing technologies;

* automotive industry, to study steering stability under adverse conditions;

* studies involving probabilistic phenomena, such as weather forecasting;

* understanding the spread of diseases;

* assessing the impact of architectural changes on computer performance;

* studying physical phenomena;

* analyzing social policies to be implemented.

## Questions Simulation Can Answer

Simulation can address many types of questions related to complex problems. Examples include:

* What...?

* How...?

* Why...?

* Can...?

## Usefulness of a Model

The usefulness of a model is directly tied to its ability to incorporate real-world elements in a simplified way.

However, a model must still include the key features of the real system, without becoming so complex that it is difficult to understand or manipulate.

This is often a difficult balance to achieve: realistic models are rarely simple, and simple models are rarely realistic.

## Models and Their Elements: Component

In mathematical models used for computer simulation, we typically identify four elements: components, variables, parameters, and functional relationships.

The component is the part of the model under simulation analysis. Model components can vary greatly depending on the system.

For example, in a network management system, components might include managed objects, management agents, and management applications.

## Models and Their Elements: Variables

Variables are used in simulation models to link one component to another.
These variables may be classified as:

* Exogenous variables: independent or input variables; may or may not be controllable.

* State variables: describe the state of a system or one of its components at the beginning, end, or during a specific time period.

* Endogenous variables: dependent or output variables of the system. These are generated from the interaction between exogenous and state variables, according to the system's operational characteristics.

## Models and Their Elements: Parameters

Exogenous variables can be treated as parameters. In this case, parameters are determined by the problem conditions or by decision-making elements.

These conditions or elements must be either read in advance as input data or generated internally if they are stochastic (via predefined methods).

## Models and Their Elements: Functional Relationships

Functional relationships describe how variables interact with model components and are of two types: identities and operational characteristics. Both are used to generate system behavior.

In a management system, for instance, traffic metrics between a management station and a remote resource could serve as an example.

An operational characteristic is a hypothesis, often expressed as a mathematical equation, that relates the system's endogenous and state variables to its exogenous variables.

## Model Construction

It's important to remember that a model is, by definition, a static representation of a system.

Simulation adds a temporal aspect to the model by describing how the system changes over time.

Simulation can be seen as the process of running experiments on a model to discover how the real system would behave under similar conditions.

In many cases, it is much easier to alter features of a simulated model than to do so in the real world (e.g., in aircraft design).


## Model Validation

Using analytical models and experimental data, researchers can employ simulation as a way to validate either the models or the data.

To test a model, results from the analytical model (when it exists) can be compared with those obtained via simulation.

Likewise, simulation can help validate experimental data, especially considering that experimental results often vary due to measurement errors, data collection variance, or errors in setting up the experiment.




## Statistical Analysis

Once a given simulation is validated, it can be used to conduct experiments that yield useful statistics about potential output results.

A researcher can design a set of experiments by identifying factors that may influence real-world results and observing whether the simulation results are statistically consistent with the experimental data.

## Types of Simulations
There are various proposed classifications for simulation models.

One common way is to divide simulations into continuous simulations and discrete-event simulations.

## Continuous Simulation

Continuous simulation is defined as a system described by state variables that change continuously over time.

State variables are system parameters that describe its behavior.
For example, consider the movement of an object, such as a ball dropped from a tall tower. One variable describing this motion is velocity.

Continuous simulation is often employed to model physical systems or systems comprising mechanical, electrical, thermal, or hydraulic components. However, it can be used to represent any system with continuous variation over time.

For example, the spread of influenza can be represented this way because, as soon as it is introduced into a population, the number of infected people begins to increase. Additionally, as the population recovers and becomes immune, the number of infected individuals decreases.

In order to represent continuous simulations on a digital computer, we use mathematical models, which consist of one or more equations that include the system's state variables.

Differential equations are the most precise form of these equations. However, they require an understanding of calculus, so we may use an algebraic substitute called a difference equation.


## Difference Equation

Difference equations allow the use of basic algebra—or even spreadsheets—to simulate a continuous system.

Using the ball example, the difference equation that defines its velocity is:

$$ v(t + Δt) = v(t) + a(t) Δt $$

Here, v(t+Δt) is the velocity at time t+Δt; v(t) is the velocity at time t and a(t) is the acceleration at time t.

Since the difference equation uses discrete time steps, it is easy to compute the next value. As the step size becomes smaller, a continuous curve of the state variable is approximated.

Still in this example, if the acceleration value is assumed to be constant, future velocity can be calculated by adding the product a(t)Δt (the area under the curve) to the current velocity.

## Error

Now, suppose the acceleration is not constant. We can still approximate the change in acceleration using a rectangle $a \times t$.

But this rectangle now introduces some error, making it impossible to calculate acceleration exactly.

However, the smaller the value of Δt, the smaller the value of the associated error.

This method is known as Euler’s method and can be used to estimate instantaneous acceleration.

## Example 3: Falling Ball
The graph clearly shows the continuous nature of the ball's velocity.

````
import numpy as np
import matplotlib.pyplot as plt # for the graphics

N = 15
Delta = 0.1
A = 9.8
v0 = 0
v = np.zeros(N)
tempo = np.zeros(N)

for t in range(1,N):
    tempo[t] = t*Delta
    v[t] = v[t-1] + A*Delta

#  Figure
fig = plt.subplots()
plt.rcParams["figure.figsize"] = (8,6)
plt.rcParams["font.size"] = "12"
plt.plot(tempo, v,"bo-");
plt.xlabel("Time (s)",fontsize=14);
plt.ylabel("Velocity (m/s)",fontsize=14);
plt.grid()
plt.show()
````

## Example 4: Biological System

When modeling the spread of a disease, we typically consider three groups of people: those susceptible to the disease (S), those infected (I), and those who have recovered (R).

Other important terms include  infection rate (λ) and the recovery rate (δ).

Given these variables, we can write the system's differential equations:

$$ \begin{equation}
   \left\{
   \begin{array}{rl}
   \frac{dS}{dt} & = & -\lambda S I  \\
   \frac{dI}{dt} & = & \lambda S I - \delta I \\
   \frac{dR}{dt} & = & \delta I
   \end{array}
   \right.
   \end{equation} $$  

From these, we can derive the corresponding difference equations to simulate the biological system:

$$ \begin{equation}
   \left\{
   \begin{array}{rl}
   S(t + \Delta t) & = & S(t) - \lambda S(t)I(t)\Delta t \\
   I(t + \Delta t) & = & I(t) + (\lambda S(t)I(t) - \delta I(t)) \Delta t \\
   R(t + \Delta t) & = & R(t) + \delta I(t) \Delta t
   \end{array}
   \right.
   \end{equation} $$

The goal is to simulate the spread of the Hong Kong flu, which affected New York during the winter of 1968–1969.

The city had a population of 7.9 million. The disease initially infected 10 people. The illness lasted about 3 days, giving a recovery rate of 0.33. The infection rate is estimated at 2.5x10⁻⁷ per day. The time step used is 1 day.

This example is a bit more complex than the falling ball since it is modeled by three dependent equations.

````
import numpy as np
import matplotlib.pyplot as plt

N = 16
S = np.zeros(N); S[0] = 7.9e6
I = np.zeros(N); I[0] = 10
R = np.zeros(N)
Tempo = np.zeros(N)
D = 1; L = 2.5e-7; delta = 0.33

for t in range (1,N):
    S[t] = S[t-1] - L*S[t-1]*I[t-1]*D
    I[t] = I[t-1] + (L*S[t-1]*I[t-1]-delta*I[t-1])*D
    R[t] = R[t-1] + delta*I[t-1]*D
    Tempo[t] = Tempo[t-1] + D

# Figure
fig = plt.subplots()
plt.rcParams["figure.figsize"] = (8,6)
plt.rcParams["font.size"] = "12"
plt.plot(Tempo, [x/1e6 for x in S],"bo-", label="Susceptible")
plt.plot(Tempo, [x/1e6 for x in I],"r^-", label="Infected")
plt.plot(Tempo, [x/1e6 for x in R],"ks-", label="Recovered")
plt.xlabel("Time (days)",fontsize=14)
plt.ylabel("Population (millions)",fontsize=14)
plt.legend(prop={"size":14},frameon=False)
plt.grid()
plt.axis([0, 15, 0, 8])
plt.show()
````

## Example 5: Falling Ball (Revisited)

In our first version, we ignored the effect of air resistance, which tends to decrease the acceleration of a falling object.

Mathematically, the system can be described by the equations below, where $F$ is the net force, $W$ is the weight, and $D$ is the drag force from the air:

$$F=W-D$$

Assuming $m$ is the object's mass, $k$ the drag coefficient, and $g$ the gravitational acceleration:

$$ma=mg-kv^2$$

The rest of the example implements this model with two different time steps and shows how they affect the simulation.



````
import numpy as np
import matplotlib.pyplot as plt

N = 80
Delta = 0.1
m = 1
k = 0.05
g = 9.8
v1 = np.zeros(N)
a1 = np.zeros(N); a1[0] = g
tempo1 = np.zeros(N)

for t in range(1,N):
    tempo1[t] = tempo1[t-1] + Delta
    v1[t] = v1[t-1] + a1[t-1] * Delta
    a1[t] = (m * g - k * v1[t]**2)/m

N = 17
Delta = 0.5
v2 = np.zeros(N)
a2 = np.zeros(N); a2[0] = g
tempo2 = np.zeros(N)

for t in range (1,N):
    tempo2[t] = tempo2[t-1] + Delta
    v2[t] = v2[t-1] + a2[t-1] * Delta
    a2[t] = (m * g - k * v2[t]**2)/m

# Figure
fig = plt.subplots()
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams["font.size"] = "12"
plt.plot(tempo1, v1,"bo-",label="$\Delta t = 0.1$")
plt.plot(tempo2, v2,"r^-",label="$\Delta t = 0.5$")
plt.xlabel("Time (s)",fontsize=14);
plt.ylabel("Velocity (m/s)",fontsize=14);
plt.legend(prop={"size":14},frameon=False)
plt.grid()
plt.show()
````
